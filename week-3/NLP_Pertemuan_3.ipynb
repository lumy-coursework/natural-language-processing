<h4> TF.IDF-Term Weighting </h4>
<ul>
    <li> Pada representasi vektor sebelumnya nilai antar term dengan dokumen hanya berupa frekuensi kemunculan saja </li>
    <li> Nilai antar term dengan dokumen dapat digantikan dengan bobot TF.IDF </li>
    <li> TF (Term Frequency) menghitung frekuensi dari suatu kata pada dokumen; sementara IDF (Inverse Document Frequency) menghitung nilai informatif dari suatu kata berdasarkan tingkat kemunculannya antar dokumen </li>

</ul>

```python
import pickle
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
PICKLE_PATH = "data/20newsgroups.pickle"

with open(PICKLE_PATH, "rb") as f:
    obj = pickle.load(f)
```

```python
if hasattr(obj, "data"):
    corpus = obj.data
elif isinstance(obj, dict):
    possible_keys = ["data", "texts", "documents", "corpus"]
    found = None
    for k in possible_keys:
        if k in obj:
            found = k
            break
    if found is None:
        raise ValueError(
            f"Pickle berbentuk dict, tapi key corpus tidak ditemukan. "
            f"Available keys: {list(obj.keys())}"
        )
    corpus = obj[found]
elif isinstance(obj, (list, tuple, np.ndarray, pd.Series)):
    corpus = obj
else:
    raise ValueError(f"Tipe objek pickle tidak dikenali: {type(obj)}")

corpus = ["" if x is None else str(x) for x in corpus]
```

```python
n_docs = len(corpus)
n_empty_raw = sum(1 for doc in corpus if doc.strip() == "")

print("Jumlah dokumen:", n_docs)
print("Jumlah dokumen kosong (raw):", n_empty_raw)
print()
```

```python
# VSM BINARY
vec_binary = CountVectorizer(binary=True)
X_binary = vec_binary.fit_transform(corpus)

# VSM TF
vec_tf = CountVectorizer()
X_tf = vec_tf.fit_transform(corpus)

# VSM TF-IDF
vec_tfidf = TfidfVectorizer()
X_tfidf = vec_tfidf.fit_transform(corpus)
```

```python
print("Shape VSM Binary :", X_binary.shape)
print("Shape VSM TF     :", X_tf.shape)
print("Shape VSM TFIDF  :", X_tfidf.shape)
print()

same_shape = (X_binary.shape == X_tf.shape == X_tfidf.shape)
print(same_shape)
```

```python
zero_rows_binary = np.sum(X_binary.getnnz(axis=1) == 0)
zero_rows_tf = np.sum(X_tf.getnnz(axis=1) == 0)
zero_rows_tfidf = np.sum(X_tfidf.getnnz(axis=1) == 0)

print()
print("Jumlah baris all-zero pada Binary :", zero_rows_binary)
print("Jumlah baris all-zero pada TF     :", zero_rows_tf)
print("Jumlah baris all-zero pada TFIDF  :", zero_rows_tfidf)

print()
print("Jumlah fitur Binary :", len(vec_binary.get_feature_names_out()))
print("Jumlah fitur TF     :", len(vec_tf.get_feature_names_out()))
print("Jumlah fitur TFIDF  :", len(vec_tfidf.get_feature_names_out()))

print()
print("Contoh 20 fitur pertama Binary:")
print(vec_binary.get_feature_names_out()[:20])
```

#### Custom TFIDF
* Menurut http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html
* default formula tf-idf yang digunakan sk-learn adalah:
* $tfidf = tf * log(\frac{N}{df+1})$ ==> Smooth IDF
* namun kita merubahnya menjadi:
* $tfidf = tf * log(\frac{N}{df})$ ==> Non Smooth IDF
* $tfidf = (1+log(tf)) * log(\frac{N}{df+1})$ ==> sublinear_tf, Smooth IDF
* $tfidf = (1+log(tf)) * log(\frac{N}{df})$ ==> sublinear_tf, Non Smooth IDF

```python
corpus = [
'This is the first document.',
'This document is the second document.',
'And this is the third one.',
'Is this the first document?',
]
tfidf_vectorizer = TfidfVectorizer(smooth_idf= False, sublinear_tf=True, lowercase=True)
tfidf = tfidf_vectorizer.fit_transform(corpus)
print(tfidf.shape) # Sama
print(tfidf[0].data) # Hanya data ini yg berubah
print(tfidf[0].indices) # Letak kolomnya = tfidf
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
# VSM awal tanpa filtering
cv_raw = CountVectorizer()
X_raw = cv_raw.fit_transform(corpus)

# VSM hasil filtering
# min_df=2  -> buang term yang muncul di < 2 dokumen
# max_df=0.8 -> buang term yang muncul di > 80% dokumen
cv_filtered = CountVectorizer(min_df=2, max_df=0.8)
X_filtered = cv_filtered.fit_transform(corpus)

# TF-IDF hasil filtering
tfidf_filtered = TfidfVectorizer(min_df=2, max_df=0.8)
X_tfidf_filtered = tfidf_filtered.fit_transform(corpus)

print("--- SHAPE SEBELUM DAN SESUDAH FILTERING ---")
print("CountVectorizer tanpa filtering :", X_raw.shape)
print("CountVectorizer dengan filtering:", X_filtered.shape)
print("TFIDF dengan filtering          :", X_tfidf_filtered.shape)
```

```python
raw_terms = np.array(cv_raw.get_feature_names_out())
filtered_terms = set(cv_filtered.get_feature_names_out())

df_raw = np.asarray((X_raw > 0).sum(axis=0)).ravel()

too_common_mask = df_raw > (0.8 * len(corpus))
too_rare_mask = df_raw < 2

too_common_terms = raw_terms[too_common_mask]
too_rare_terms = raw_terms[too_rare_mask]

removed_terms = [t for t in raw_terms if t not in filtered_terms]
```

```python
print("\n--- RINGKASAN FILTERING ---")
print("Jumlah dokumen                    :", len(corpus))
print("Jumlah term sebelum filtering     :", len(raw_terms))
print("Jumlah term sesudah filtering     :", len(filtered_terms))
print("Jumlah term dibuang               :", len(removed_terms))
print("Jumlah term terlalu sering muncul :", len(too_common_terms))
print("Jumlah term terlalu jarang muncul :", len(too_rare_terms))

print("\nContoh term terlalu sering muncul:")
print(too_common_terms[:20])

print("\nContoh term terlalu jarang muncul:")
print(too_rare_terms[:20])

print("\nContoh term yang tersisa setelah filtering:")
print(sorted(list(filtered_terms))[:30])
```

Jawaban:

1. Menghilangkan term dengan frekuensi kemunculan tinggi antar dokumen perlu dilakukan
   karena term seperti ini biasanya kurang diskriminatif. Term tersebut muncul di terlalu
   banyak dokumen sehingga kurang membantu membedakan topik satu dokumen dari dokumen lain.

2. Menghilangkan term dengan frekuensi kemunculan rendah antar dokumen juga penting
   karena term yang sangat jarang sering bersifat noise, typo, atau terlalu spesifik,
   sehingga kontribusinya ke representasi umum corpus kecil tetapi menambah dimensi fitur.

3. Filtering berdasarkan frekuensi perlu dilakukan agar ukuran vocabulary lebih ringkas,
   representasi lebih bersih, dan proses komputasi lebih efisien.
