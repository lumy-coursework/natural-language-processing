**Clean and tokenize function**

```python
def tokenize_clean(text):
    
    #tokenisasi
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word
        in nltk.word_tokenize(sent)]
    
    #clean token from numeric and other character like puntuation
    filtered_tokens = []
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
            
    return filtered_tokens
```

<br>

### Step 01 - Get Indonesian stop words

```python
stopwords = nltk.corpus.stopwords.words('indonesian')
```

```python
len(stopwords)
```

<br>

### Step 02 - Create remove_stopwords function

```python
def remove_stopwords(tokenized_text):
    cleaned_token = []
    
    for token in tokenized_text:
        if token not in stopwords:
            cleaned_token.append(token)
    
    return cleaned_token
```

### Step 03 - Create dataset

```python
string01 = "Perekonomian Indonesia sedang dalam pertumbuhan yang membanggakan dan masyarakat menikmati hasil pembangunan."
string02 = "Mahasiswa sedang belajar pemrosesan bahasa alami dan mengerjakan tugas stopwords removal di kelas."

dataset = [string01, string02]

dataset
```

### Step 04 - Process string01

```python
print("Teks asli:")
print(string01)

print("\nHasil tokenize_clean:")
tokens01 = tokenize_clean(string01)
print(tokens01)

print("\nHasil remove_stopwords:")
print(remove_stopwords(tokens01))
```

### Step 05 - Process string02

```python
print("Teks asli:")
print(string02)

print("\nHasil tokenize_clean:")
tokens02 = tokenize_clean(string02)
print(tokens02)

print("\nHasil remove_stopwords:")
print(remove_stopwords(tokens02))
```
