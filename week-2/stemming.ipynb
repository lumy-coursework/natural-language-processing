#### Import required library

```python
import nltk
import re
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

#### Clean and tokenize function

```python
def tokenize_clean(text):
    
    #tokenisasi
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word
        in nltk.word_tokenize(sent)]
    
    #clean token from numeric and other character like puntuation
    filtered_tokens = []
    for token in tokens:
        if re.search('[a-zA-Z]', token):
            filtered_tokens.append(token)
            
    return filtered_tokens
```

<br>

#### Get Indonesian stop words

```python
stopwords = nltk.corpus.stopwords.words('indonesian')
```

<br>

#### Remove stopwords function

```python
def remove_stopwords(tokenized_text):
    
    cleaned_token = []
    for token in tokenized_text:
        if token not in stopwords:
            cleaned_token.append(token)
            
    return cleaned_token
```

### Step 01 - Create stemming_text Function

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming_text(text):
    tokens = tokenize_clean(text)
    
    tokens_no_stop = remove_stopwords(tokens)
    
    stemmed_tokens = [stemmer.stem(token) for token in tokens_no_stop]
    
    return stemmed_tokens
```

### Step 02 - Create Dataset

```python
string01 = "Isi File: Sangat disarankan untuk menyertakan file README.md pada commit pertama agar orang lain (atau kamu di masa depan) tahu proyek apa ini."
string02 = "Gunakan Imperative Mood: Selalu gunakan kata kerja dasar/perintah (misalnya add bukan added, init bukan initialized)"

data = pd.DataFrame({
    "raw_text": [string01, string02]
})

data
```

### Step 03 - Process Data and View The Result

```python
data["tokens_clean"] = data["raw_text"].apply(tokenize_clean)
data["no_stopwords"] = data["tokens_clean"].apply(remove_stopwords)
data["stemmed_tokens"] = data["raw_text"].apply(stemming_text)
data["stemmed_text"] = data["stemmed_tokens"].apply(lambda x: " ".join(x))

data
```

```python
for i, row in data.iterrows():
    print(f"Data {i+1}")
    print("Raw text       :", row["raw_text"])
    print("Tokens clean   :", tokenize_clean(row["raw_text"]))
    print("No stopwords   :", remove_stopwords(tokenize_clean(row["raw_text"])))
    print("Stemmed tokens :", stemming_text(row["raw_text"]))
    print("Stemmed text   :", " ".join(stemming_text(row["raw_text"])))
    print("-" * 60)
```

### Step 04 - Process string02

```python
print("Teks asli:")
print(string02)

print("\nHasil tokenize_clean:")
print(tokenize_clean(string02))

print("\nHasil remove_stopwords:")
print(remove_stopwords(tokenize_clean(string02)))

print("\nHasil stemming_text:")
print(stemming_text(string02))

print("\nHasil akhir:")
print(" ".join(stemming_text(string02)))
```
